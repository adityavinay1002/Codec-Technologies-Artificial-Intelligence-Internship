```python
# Spam Email Classifier (User Input Version)

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Sample Dataset
data = {
    'text': [
        'Win money now',
        'Hello how are you',
        'Claim your free prize',
        'Meeting at 5 pm',
        'Congratulations you won lottery',
        'Lets have lunch tomorrow'
    ],
    'label': [1,0,1,0,1,0]   # 1 = Spam, 0 = Not Spam
}

df = pd.DataFrame(data)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['label'], test_size=0.3, random_state=42
)

# Convert text to numerical form
cv = CountVectorizer()
X_train_cv = cv.fit_transform(X_train)
X_test_cv = cv.transform(X_test)

# Train model
model = MultinomialNB()
model.fit(X_train_cv, y_train)

# Accuracy
predictions = model.predict(X_test_cv)
print("Model Accuracy:", accuracy_score(y_test, predictions))

# User input after running
user_mail = input("Enter Email Message: ")

# Predict entered mail
mail_vector = cv.transform([user_mail])
result = model.predict(mail_vector)

if result[0] == 1:
    print("This is a Spam Email")
else:
    print("This is Not a Spam Email")
```
